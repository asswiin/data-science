```python
import matplotlib.pyplot as pyp
import numpy as np
pyp.suptitle("GRAPH")
x=np.array([1,2])
y=np.array([1,4])
pyp.subplot(3,3,1)
pyp.title("Title 1")
pyp.xlabel("blue color",c="b")
pyp.ylabel("blue color",c="b")
pyp.scatter(x,y)
pyp.tight_layout()


x=np.array([5,10,5,10,5])
y=np.array([9,6,9,6,9])
pyp.subplot(3,3,2)
pyp.plot(x,marker="*",c="orange",lw="2.5")
pyp.plot(y,ls="dashed")
pyp.tight_layout()

x=np.array([5,10,5,10])
pyp.subplot(3,3,3)
pyp.plot(x,marker="o",ms="9",mfc="yellow",mec="black")
pyp.tight_layout()

x=np.array([5,10,15,20])
pyp.subplot(3,3,4)pyp.subplot(3,3,7)
pyp.grid()                     
pyp.plot(x)
pyp.tight_layout()

x=np.array([1,4,5,7,10])
y=np.array([2,3,8,9,3])
pyp.subplot(3,3,5)
colors = np.array(["red","green","blue","yellow","pink"])
sizes=np.array([20,40,60,80,120])
pyp.scatter(x,y,c=colors,s=sizes,alpha=0.5)
pyp.plot()
pyp.tight_layout()

x = np.array(["A", "B", "C", "D"])
y = np.array([5,3,8,2])
pyp.subplot(3,3,6)
pyp.bar(x,y,color='r',width = 0.1)
pyp.show()

x = np.array([25,35,20,20])

lab=["Explode","shadow","color","legend"]

exp=[0.2,0,0,0]
pyp.pie(x,labels=lab,explode=exp)
pyp.legend(loc="upper right")

pyp.show()
```
