```python
# mean,median,mode

import pandas as pd

df = pd.read_csv('data.csv')

x=df["Calories"].mean()
x=df["Calories"].median()
x=df["Calories"].mode()[0]


df.fillna({"Calories":x}, inplace=True)

print(df.head(7).to_string())
```
